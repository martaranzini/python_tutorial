# Working with files

```python
# We want to store all the filenames into a list and then iterate to create plots of each file
```

```python
import glob # that will help us obtain the files from a directory
print(glob.glob('data/inflammation*.csv'))
```

```python
%matplotlib inline

# we import the tools that we need
import numpy
import matplotlib.pyplot

filenames = glob.glob('data/inflammation*.csv')

for f in filenames[:3]:

    # We load the inflammation into data
    data = numpy.loadtxt(fname=f, delimiter=",")

    # Create a canvas that has size 10 in rows and 3 in columns
    fig = matplotlib.pyplot.figure(figsize=(10, 3))
    # Create the axes (three subplots)
    axes1 = fig.add_subplot(1,3,1)
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)

    axes1.set_ylabel("average")
    axes1.plot(data.mean(axis=0))

    axes2.set_ylabel("max")
    axes2.plot(data.max(axis=0))

    axes3.set_ylabel("min")
    axes3.plot(data.min(axis=0))
    axes3.set_ylim(0,6)

    matplotlib.pyplot.show
```
